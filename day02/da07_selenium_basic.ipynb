#### 크롬 드라이버 설치
1. 현재 크롬 버전에 맞는 버전에 맞는 크롬드라이버 압축파일 다운로드
2. 특정 폴더 압축해제
3. 시스템속성 > 고급 > 환경변수 등록(sysdm.cpl)
    - path 맨 마지막에 크롬 드라이버 경로를 추가
4. 명령 프롬프트
    ```shell
    > chromedriver --version
    ```

````python
from selenium import webdriver
````

````python
## 웹드라이버에서 크롬을 실행
driver = webdriver.Chrome()
````

````python
## 제어할 수 있는 크롬에서 조작을 시작
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
````

````python
html = driver.page_source
````
