````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
## 단일차트
plt.plot([1, 2, 3, 4, 5])
plt.show()    # [<matplotlib.lines.Line2D at 0x218882de090>]와 같은 객체 정보 출력이 사라짐
````

````python
plt.plot(np.random.randn(50))   # np.random.randn(n) n개의 정수를 랜덤하게 만들어줌
plt.show()
````

````python
## 다중차트, figure()사용
fig = plt.figure()
plt1 = fig.add_subplot(2, 2, 1) # 1번 차트(하나의 그림에)
plt2 = fig.add_subplot(2, 2, 2) # 2번 차트
plt3 = fig.add_subplot(2, 2, 3)
plt4 = fig.add_subplot(2, 2, 4)

plt1.plot(np.random.randn(50))
plt2.hist(np.random.randn(100))
plt3.scatter(np.arange(30), np.random.randn(30))
plt4.plot(np.random.randn(30), 'g') # 색상 'r', 'g', 'b', 'y', 'v'

plt.show()
````

````python
## 다중차트 두 번째, 4개 이상 가능
# subplots() figsize로 사이즈 조절 가능
fig, axes = plt.subplots(2, 3, figsize=(10, 6))
axes[0][0].plot(np.random.randn(50))
axes[0][1].hist(np.random.randn(100))
axes[1][0].scatter(np.arange(30), np.random.randn(30))
axes[1][1].plot(np.random.randn(30), 'g')

plt.show()
````

#### 차트 스타일 지정 옵션
- label, stylem alpha, rot, xticks, yticks, xlim, ylim, color, marker, ...


````python
fig, axes = plt.subplots(2, 3, figsize=(10, 6))
# color = 'k' 검은색, 'crimson'핏빛
# alpha = 0.0 완전 투명, 1.0 불투명
# bins = 그루핑 개수
# linestyle = 선 종류 solid, dotted, dashed, dashdot, -, :, --, -.
# marker = 점 종류 *(별), ^(삼각형), o(동그라미)
axes[0][0].plot(np.random.randn(50), color='k', alpha=0.2)
axes[0][1].hist(np.random.randn(100), color='g', bins=10)
axes[0][2].plot(np.random.randn(50), color='k', linestyle='--')
axes[1][0].scatter(np.arange(30), np.random.randn(30), marker='*')
axes[1][1].plot(np.random.randn(30), 'g')
axes[1][2].plot(np.random.randn(20), marker='o')

plt.show()
````

````python
# cumsum() sumulative sum(누적합)
# legend() 범례 표시함수
# 플롯에 데이터가 여러개 표시될 때 파랑, 주황, 초록 순
# 위치 값 : 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 
#           'center left', 'center right', 'lower center', 'upper center', 'center'
plt.plot(np.random.randn(50).cumsum(), label='data1')
plt.plot(np.random.randn(50).cumsum(), label='data2')
plt.plot(np.random.randn(50), label='data3')   # cumsum 생략 가능
plt.legend(loc='upper left')

plt.show()
````

````python
## x, y축 설정
plt.figure(figsize=(18, 5))
plt.plot(np.random.randn(1000))
# y축과 개수 동일하게
plt.xticks([0,250, 500, 750, 1000], labels=['min', '25%', 'median', '75%', 'max'])
plt.title('Stastics')   # 한글 불가

plt.show()
````

#### Matplotlib 한글 표시 문제
- 기본적으로 한글을 지원. font_manager 객체를 사용해서 한글을 쓸 수 있도록 변경
- 한글을 기본으로 사용하기 때문에 항상 이 아래 부분은 노트북을 사용할 때 마다 복사해서 사용

````python
from matplotlib import rcParams, font_manager, rc

font_path = 'C:/Windows/Fonts/NanumGothicCoding.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()  # 실제 설치된 폰트 이름조회
rc('font',family=font)  # 한글 깨짐 현상 해결
rcParams['axes.unicode_minus'] = False  # 한글 사용 시 마이너스 깨짐 해결
````

````python
# seaborn 스타일 지정 ['white', 'dark', whitegrid', 'darkgrid', 'ticks']
sns.set_style('darkgrid')
````

````python
flights[['month', 'passengers']].plot()
plt.show()
````

````python
sns.boxplot(data=flights, x='year', y='passengers')
plt.show()
````

````python
sns.swarmplot(data=flights, x='year', y='passengers')
plt.show()
````

````python
sns.lineplot(data=flights, x='year', y='passengers')
plt.show()
````
