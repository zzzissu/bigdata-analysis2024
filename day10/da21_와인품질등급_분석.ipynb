##### statsmodels 라이브러리 분석 함수
- OLS : Ordinary Least Square 2승 모델로 분석을 하는 함수
- GLM : Generalized Linear Model 일반 선형모델로 분석 함수
- GLS : Generalized Least Square 2승 일반화된 최소 승수 모델(두가지로 표본집단을 나눠서)
- WLS : Weighted Least Square 가중치를 부가한 OLS 모델

- fit() : 훈련결과를 내는 함수
- predict() : 결과를 예측하는 함수

```python
## 필수 라이브러리/모듈 사용등록
import pandas as pd
from ucimlrepo import fetch_ucirepo # 인터넷으로 통계 데이터를 다운로드함 함수
from scipy import stats # 과학수치모듈 내의 기술통계 객체 사용
from statsmodels.formula.api import ols, glm, wls, gls    # 분석을 위한 결과를 도출해주는 함수들
```

```python
## 와인 데이터 가져오기
wineQuality = fetch_ucirepo(id=186)
```

```python
dfWineTotal = wineQuality.data.original
```

```python
dfWineTotal.describe()
```

#### 회귀분석

```python
## 회귀분석 변수로 회귀수식 작성
Rfotmula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar	+ chlorides	+ free_sulfur_dioxide + total_sulfur_dioxide + density + pH	+ sulphates	+ alcohol'
```

```python
## 회귀분석
regrResult = ols(Rfotmula, data=dfWineTotal).fit()
regrResult.summary()
```

```python
regrResult2 = gls(Rfotmula, data=dfWineTotal).fit()
regrResult2.summary()
```

##### 도출된 회귀분석 결과로 새로운 와인샘플의 품질등급 예측

```python
## 와인 품질등급 예측용 샘플데이터
dfWineTotal.columns.difference(['quality', 'color'])
```

```python
# 품질예측할것이기 때문에 quality 불필요, 와인종류도 의미없음
sample1 = dfWineTotal[dfWineTotal.columns.difference(['quality', 'color'])]
```

```python
sample1 = sample1[0:5]
sample1
```

```python
dfWineTotal.head()
```

```python
# OLS 모델 예측시작
regrResult.predict(sample1)
```

```python
## GLM 모델 예측
regrResult2.predict(sample1)
```
