```python
# 셀레니움 라이브러리 사용등록
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
```

#### BeautifulSoup4 사용

```python
# BeautifulSoup4 사용등록
from bs4 import BeautifulSoup
```

```python
html = driver.page_source   # 웹 드라이버로 열고 있는 페이지의 소스
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

#### BeautifulSoup4
- html로 파싱된 html 소스 코드를 태그별로 필요한 것만 가지고 오는 처리

```python
# html은 부모, 자식의 관계로 모두 구성
# 내가 필요한 자식 안에 데이터를 가지고 올려면 부모까지 지정해줘야 함
top100 = soup.select('table > tbody > tr')
len(top100)
```

```python
top100[0]
```

```python
song1 = top100[0]
```

```python
## select 함수를 잘 활용해서 필요한 정보를 찾을 수 있음
title = song1.select('a')
len(title)
```

```python
## title (배열) 인덱스로 필요한 정보까지 찾을 수 있음
title[3]
```

```python
## select 함수를 여러번 테스트 해서 찾고 싶은 정보를 확인
title = song1.select('span > a')
```

```python
title[0].text
```

```python
title[1].text
```

#### bs4 필요한 정보 가져오기

```python
songRanking = []
num = 1     # 멜론차트 100위까지 추출
for song in top100:
    rank = num
    title = song.select('span > a')[0].text
    singer = song.select('span > a')[1].text
    num += 1
    print(rank, title, singer)
    songRanking.append([rank, title, singer])
```

```python
import pandas as pd
```

```python
df_top100 = pd.DataFrame(songRanking, columns=['순위', '노래제목', '가수'])
```

```python
df_top100.to_excel('./data/melon_top100.xlsx', index=False)
```
