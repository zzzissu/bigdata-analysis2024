```python
## 필요 라이브러리 사용등록
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
```

```python
# 데이터 준비
bCancer = load_breast_cancer()
```

```python
bCancer.feature_names
```

```python
len(bCancer.data)
```

```python
bCancer.target
```

```python
## DF 생성
dfBrCancer = pd.DataFrame(bCancer.data, columns=bCancer.feature_names)
```

```python
dfBrCancer
```

```python
# 종속변수 암판단결과 추가
dfBrCancer['result'] = bCancer.target
```

```python
dfBrCancer
```

```python
dfBrCancer.isnull().sum()
```

```python
dfBrCancer.info()
```

```python
## 사이킷런 필요 라이브러리 사용등록
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
## 특성으로 사용할 데이터의 평균이 0, 분산이 1이 되는 정규 분포형태로 스케일링
## 로지스틱 회귀 모델에서 X축으로 사용하기 위해서
brCancerScaled = scaler.fit_transform(bCancer.data)
```

```python
bCancer.data[0]
```

```python
brCancerScaled[0]
```

##### 로지스틱 회귀로 분석모델 구축
- LogisticRegression

```python
## 분석용 라이브러리 사용등록
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
## x(독립변수), y(종속변수 / 악성|양성 결과값)
y = dfBrCancer['result']
X = brCancerScaled
```

```python
# 훈련용데이터와 검증(평가)용 데이터로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
len(X_test)
```

```python
## 로지스틱 회귀 분석 모델생성
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
# 예측 시작
y_predict = model.predict(X_test)
```

##### 분석평가지표로 실제결과/예측결과 비교
- 정확도, 정밀도, 재현율, F1, ROC_AUC_Score

```python
## 평가지표용 라이브러리 사용등록
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
## 1. 오차행렬(실제결과, 예측결과의 오차행렬 결과)
## 171개 확인결과 0을 0으로 예측한 건 60건,
## 0을 1로 잘못 예측한 건 3건,
## 1을 1로 예측한 건 107건
confusion_matrix(y_test, y_predict)
```

```python
## 정확도부터 ROC 기반 AUC스코어 까지
accuracy = accuracy_score(y_test, y_predict)
precision = precision_score(y_test, y_predict)
recall = recall_score(y_test, y_predict)
f1 = f1_score(y_test, y_predict)
roc_auc = roc_auc_score(y_test, y_predict)
```

```python
f'정확도: {accuracy:.3f}, 정밀도: {precision:.3f}, 재현율: {recall:.3f}, F1: {f1:.3f}, ROC_AUC: {roc_auc:.3f}'
```

##### 결론
- 1/0으로 참, 거짓 등의 값을 판단할 때는 로지스틱회귀를 사용할 것
